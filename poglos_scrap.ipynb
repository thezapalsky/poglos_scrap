```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import re
import time
```

```python
def get_url(url):
    try:
        r = requests.get(url)
    except Exception as e:
        print(e)
        return None
    else:
        # do something if no exception
        #print(f"Status code:  {r.status_code}")
        #print(f"Content type: {r.headers['content-type']}")
        # print(r.text[:1000])
        return r.text
```

```python
url_clean='https://www.polskieradio.pl/10/6071/Strona/'
urls = []
secs = 1

for page_nr in range(30):
    time.sleep(secs)
    bs = BeautifulSoup(get_url(url_clean+str(page_nr)), 'lxml')
    for link in bs.find_all('a'):
        try:
            if('/10/6071/Artykul' in link.get('href') ):
                urls.append(link.get('href'))
        except:
            pass
```

```python
links = []

for url in urls:
    try:
        #time.sleep(secs)
        #url = 'https://www.polskieradio.pl/10/6071/Artykul/2207126,Poglos-25-pazdziernika-godz-1905'
        url = 'https://www.polskieradio.pl' + url
        bs = BeautifulSoup(get_url(url), 'html.parser')
        obj = bs.find(id="box-sounds").script
        for string in obj.stripped_strings:
            jquery_string = repr(string)
        m = re.search('static.*mp3', jquery_string)
        link = m.group(0)
        links.append(link)
    except:
        links.append('exception')
```

```python
df = pd.DataFrame( list(zip(urls, links)),columns=['url','mp3'] )
df.to_csv('poglos.csv',index=True)
```
